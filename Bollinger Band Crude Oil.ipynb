```python
import pandas as pd  
import numpy as np   
import datetime
import matplotlib.pyplot as plt  
```

```python
data=pd.read_csv('Combined_Data_ADS_Index.csv').dropna()
data=data.iloc[::,:2]  
data.columns=['observation_date','CRUDE OIL Actual Price'] 
data
```

```python
data['moving average']=data['CRUDE OIL Actual Price'].rolling(26).mean()  
data['standard deviation']=data['CRUDE OIL Actual Price'].rolling(20).std()  
data['upper band']=data['moving average']+data['standard deviation']  
data['lower band']=data['moving average']-data['standard deviation'] 
data 
```

```python
fig = plt.figure(figsize=(15,8))
plt.plot(data['CRUDE OIL Actual Price'],'y',label = "CRUDE OIL Actual Price")
plt.plot(data['moving average'],'b--',label = "moving average")
plt.plot(data['upper band'],'r',label = "upper band")
plt.plot(data['lower band'],'g',label = "lower band")
plt.title('$BOLLINGER$')
plt.legend()
plt.show()
```

#### Two trading bands are placed around a moving average with two standard deviations above and below average
#### Each time the price touches the lower band i.e. market bottom we should sense a buying opportunity.
